```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
file_path = 'heart_cleaned.csv'  # Change this to the correct path if needed
df = pd.read_csv(file_path)
```

```python
#  Distribution of the target variable (HeartDisease)
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='HeartDisease')
plt.title('HeartDisease Distribution')
plt.show()
```

```python
# Distribution of each variable
df.hist(figsize=(15, 10), bins=20, color='skyblue', edgecolor='black')
plt.suptitle('Distribution of Each Variable', fontsize=16)
plt.show()
```

```python
#  Heatmap of correlation between variables
plt.figure(figsize=(12, 8))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Heatmap of Correlations')
plt.show()
```
